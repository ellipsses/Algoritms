**Дано:**

*   n образцов бабочек (вершины)
*   m оценок пар образцов: «одинаковые» или «разные» (ребра)

Обозначим ребра следующим образом:
*   Ребро типа 0 – одинаковые бабочки (вершины)
*   Ребро типа 1 – разные бабочки (вершины)


**Алгоритм действий:**

Берем любую бабочку, условно относим её к виду А (обозначаем ее за 0), и начинаем проверять её соседей.

* Если сосед должен быть «таким же» — обозначаем его тем же номером (0).
*	Если сосед должен быть «другим» — обозначаем его другим номером (1).

И так алгоритм идет вглубь по цепочкам ограничений. Если натыкается на бабочку, которой уже дан номер, то сверяет: совпадает ли её номер с тем, который требует заданное утверждение:

Если нет — значит, данные противоречивы друг другу.

Если прошли весь список и конфликтов нет — предложенное разделение возможно и данные непротиворечивы друг другу.


```python
def is_consistent(n, evaluations):
    graph = [[] for i in range(n)]
    for i, j, relation in evaluations:
        edge_type = 0 if relation == 'same' else 1 # ребра переводим в числа
        graph[i].append((j, edge_type)) # пары вида (сосед, тип пары)
        graph[j].append((i, edge_type))  # граф неориентированный

    # массив с номерами вершин, -1 - не посещена, 0 или 1 - присвоенный вид
    number = [-1] * n

    # обход в глубину
    def dfs(u, current_number):
        number[u] = current_number
        for v, edge_type in graph[u]:
            # Если сосед same (0), номер вершины не меняется
            # Если сосед different (1), номер вершины инвертируется через XOR
            required_number = current_number ^ edge_type

            if number[v] == -1: # если соседа еще не посетили - вглубь
                if not dfs(v, required_number):
                    return False
            elif number[v] != required_number: # если у соседа номер не совпадает с требованием
                return False
        return True

    for v in range(n):
        if number[v] == -1:
            if not dfs(v, 0): # проверка с номера 0
                return "противоречие"
    return "противоречия нет"

n1 = 4
evals1 = [
    (0, 1, "same"),
    (1, 2, "different"),
    (2, 3, "different"),
    (0, 3, "same")
]
print(f"Кол-во образцов: {n1}, утверждения: {evals1}")
print(f"Результат: {is_consistent(n1, evals1)}\n")

n2 = 3
evals2 = [
    (0, 1, "same"),
    (1, 2, "same"),
    (0, 2, "different")
]
print(f"Кол-во образцов: {n2}, утверждения: {evals2}")
print(f"Результат: {is_consistent(n2, evals2)}")
```
